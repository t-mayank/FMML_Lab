```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'engine-size': [130, 165, 150, 120, 155, 180, 150],
    'highway-mpg': [30, 22, 24, 35, 29, 22, 26],
    'body-style': ['sedan', 'sedan', 'hatchback', 'sedan', 'hatchback', 'convertible', 'sedan'],
    'price': [15000, 18000, 12000, 17000, 16000, 30000, 20000]
}

df = pd.DataFrame(data)
X = df[['engine-size', 'highway-mpg', 'body-style']]
y = df['price']
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

sns.regplot(x=X['highway-mpg'], y=y, ax=axs[0], color='purple')
axs[0].set_title('Highway-MPG vs Price')

sns.boxplot(x=X['body-style'], y=y, palette="Set3", ax=axs[1])
axs[1].set_title('Body Style vs Price')

plt.tight_layout()
plt.show()
```

```python
numeric_features = ['engine-size', 'highway-mpg', 'price']
corr_matrix = df[numeric_features].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
